```python
import pandas as pd
from datetime import datetime, time
```

```python
df1 = pd.DataFrame.from_csv('btc.csv')
df2 = pd.DataFrame.from_csv('eth.csv')
df3 = pd.DataFrame.from_csv('xmr.csv')
df4 = pd.DataFrame.from_csv('xmrbtc.csv')
```

```python
df1.head()
```

```python
df = pd.merge(df1, df2, on='time',suffixes=('','_e'))
df = pd.merge(df, df3, on='time', suffixes=('','_m'))
df = pd.merge(df, df4, on='time', suffixes=('','_mb'))
```

```python
df
```

```python
df.drop_duplicates()
```

```python
def shift(column):
    df[column+'_diff'] = df[column]-df[column].shift(1)
    
def compare(column1, column2):
    df[column1+'_comp'] = df[column1] - df[column2]
```

```python
base = ['close','high','low','open','volumefrom','volumeto']
for b in base:
    shift(b)
    shift(b+'_e')    
    shift(b+'_m')    
    shift(b+'_mb')   
    
compare('close','open')
compare('close_e','open_e')
compare('close_m','open_m')
compare('close_mb','open_mb')

compare('high','low')
compare('high_e','low_e')
compare('high_m','low_m')
compare('high_mb','low_mb')
```

```python
def get_start(ts):
    t = datetime.fromtimestamp(ts)
    return t.hour*60 + t.minute

def get_start_hour(ts):
    t = datetime.fromtimestamp(ts)
    return t.hour

def get_day(ts):
    t = datetime.fromtimestamp(ts)
    return t.weekday()
```

```python
df['from_start'] = df['time'].apply(get_start)
df['from_start_hour'] = df['time'].apply(get_start_hour)
df['day'] = df['time'].apply(get_day)
```

```python
for b in base:
    df = df.drop(b, axis=1)
    df = df.drop(b+'_e', axis=1)    
    df = df.drop(b+'_m', axis=1)    
    df = df.drop(b+'_mb', axis=1)   

df = df.drop('time', axis=1)       
df.drop(df.index[0], inplace=True)
```

```python
pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df.to_csv('final.csv', index=False, header = False)
```
