```python
import json
import urllib.request
import time
import pandas as pd
from time import sleep
import datetime
```

```python
base_url = 'https://min-api.cryptocompare.com/'
```

```python
def get_info(time, currency, currency2='USD'):
    data = json.load(urllib.request.urlopen(base_url + f'data/histominute?fsym={currency}&tsym={currency2}&limit=60&e=CCCAGG&toTs={time}'))
    return data
```

```python

get_info('1514463240','BTC')
```

```python
ts = int(time.time())
ts = ts - (7*23*60*60)
```

```python
dfb = pd.DataFrame(columns=['close','high','low','open','time','volumefrom','volumeto'])
dfe = pd.DataFrame(columns=['close','high','low','open','time','volumefrom','volumeto'])
dfm = pd.DataFrame(columns=['close','high','low','open','time','volumefrom','volumeto'])
dfmb = pd.DataFrame(columns=['close','high','low','open','time','volumefrom','volumeto'])
```

```python
while ts < time.time():
    try:
        btc = get_info(ts,'BTC')
        eth = get_info(ts,'ETH')    
        monero = get_info(ts,'XMR')  
        monerobt = get_info(ts,'XMR','BTC')
        sleep(0.5)
        for x in range(0, 60):
            dfb = dfb.append(btc['Data'][x],ignore_index=True)
            dfe = dfe.append(eth['Data'][x],ignore_index=True)        
            dfm = dfm.append(monero['Data'][x],ignore_index=True)        
            dfmb = dfmb.append(monerobt['Data'][x],ignore_index=True)        
        ts = ts + 60*60
        dfb.to_csv('btc.csv')
        dfe.to_csv('eth.csv')    
        dfm.to_csv('xmr.csv')    
        dfmb.to_csv('xmrbtc.csv')    
        print(datetime.datetime.fromtimestamp(
            ts
        ).strftime('%Y-%m-%d %H:%M:%S'))
    except:
        pass
```
